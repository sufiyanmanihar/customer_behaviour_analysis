```python
# here we should not fill the null values of Review Rating with the median value because the the rating is based on the categories (clothing and footwear)
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Creating a column age group
labels = ['Young Adults','Adults','Middle Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Bi-Weekly' :14,
    'Quarterly' : 90,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['frequency_purchase_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_purchase_days','frequency_of_purchases']].head(10)
```

```python
df
```

```python
# Comparing if both the tables have the same values in each rows so that we can remove one of them
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```
